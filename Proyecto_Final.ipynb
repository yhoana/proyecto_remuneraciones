```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Cargo los datos:
import pandas as pd
df = pd.read_csv('https://catalogodatos.gub.uy/dataset/b8fac825-88c7-4131-b1ca-c1f96be6186b/resource/ed4ccd0e-7e7f-46ec-8e23-b1bdee45501f/download/remuneraciones.csv')
print()
```

```python
df = df.dropna(axis= 0)
df
```

```python
#  Ocupados vs Vacantes

from matplotlib import pyplot as plt
df.plot(kind='scatter', x='Ocupados', y='Vacantes', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['SueldoBasico'], bins=30, kde=True)
plt.title('DistribuciÃ³n de Remuneraciones')
plt.xlabel('SueldoBasico')
plt.ylabel('Frecuencia')
plt.show()

#Aqui hacemos un histograma
```
