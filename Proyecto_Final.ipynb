```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Cargo los datos:
import pandas as pd
df = pd.read_csv('https://catalogodatos.gub.uy/dataset/b8fac825-88c7-4131-b1ca-c1f96be6186b/resource/ed4ccd0e-7e7f-46ec-8e23-b1bdee45501f/download/remuneraciones.csv')
print()
```

```python
print(df.head())

```

```python
print(df.describe())
```

```python
df = df.dropna(axis= 0)
df
#Aqui limpio los datos:
```

```python
#SueldoBasico

from matplotlib import pyplot as plt
df['SueldoBasico'].plot(kind='line', figsize=(8, 4), title='SueldoBasico')
plt.gca().spines[['top', 'right']].set_visible(False)
```

**Aqui creamos un grafico de barras para poder visualizar la cantidad de cargos que hay segun el Escalafon:**

```python
conteo_cargos = df.groupby('Escalafon')['Cargo'].count().reset_index()

# Renombramos  columnas para mayor claridad
conteo_cargos.columns = ['Escalafon', 'Cantidad de Cargos']

# Grafico de barras
plt.figure(figsize=(20,6))
sns.barplot(x='Escalafon', y='Cantidad de Cargos', data=conteo_cargos)
plt.title('Cantidad de cargos por escalafón')
plt.xlabel('Escalafón')
plt.ylabel('Cantidad de cargos')
plt.show()
```

**Aqui vamos a generar un mapa de calor (heatmap) para poder ver donde se concentran los sueldos segun el escalafon:**

```python

pivot_table = df.pivot_table(index='Escalafon', values='SueldoBasico', aggfunc='mean')


plt.figure(figsize=(8, 6))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap='YlGnBu')

plt.title('Concentración de Sueldos Más Altos por Escalafón')
plt.xlabel('Escalafón')
plt.ylabel('')

plt.show()
```

**En el siguiente codigo, con los datos de nuestro DF, vamos a mostrar las vacantes cubiertas y No cubiertas, para poder visualizar a traves de la cantidas de plazas abiertas para cada cargo cuantos se llegan a cubrir:**

```python
df['Porcentaje Cubiertas'] = (df['Ocupados'] / df['Plazas']) * 100

total_plazas = df['Plazas'].sum()
total_ocupados = df['Ocupados'].sum()

porcentaje_cubiertas_total = (total_ocupados / total_plazas) * 100

# Grafico de barras
plt.figure(figsize=(8,6))
plt.bar(['Vacantes Cubiertas', 'Vacantes No Cubiertas'],
        [porcentaje_cubiertas_total, 100 - porcentaje_cubiertas_total],
        color=['green', 'red'])

plt.ylabel('Porcentaje (%)')
plt.title('Porcentaje de Vacantes Cubiertas')
plt.ylim(0, 100)

# Porcentajes en BaRras
plt.text(0, porcentaje_cubiertas_total + 2, f'{porcentaje_cubiertas_total:.2f}%', ha='center')
plt.text(1, 100 - porcentaje_cubiertas_total + 2, f'{100 - porcentaje_cubiertas_total:.2f}%', ha='center')

plt.show()
```

Con el siguiente codigo vamos a mostrar un Histograma, con los datos de nuestro DF de Remuneraciones:

```python
plt.figure(figsize=(10,6))
sns.histplot(df['SueldoBasico'], bins=30, kde=True)
plt.title('Distribución de Remuneraciones')
plt.xlabel('SueldoBasico')
plt.ylabel('Frecuencia')
plt.show()
```
